```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample website traffic dataset
data = {
    "Page": ["Home", "Products", "Cart", "Checkout", "About", "Contact"],

    "PageViews": [500, 350, 200, 120, 150, 100],

    "AvgSessionDuration": [5, 4, 3, 2, 2, 1],

    "BounceRate": [20, 30, 40, 50, 25, 35]
}

# Create DataFrame
df = pd.DataFrame(data)

print("Website Traffic Dataset:\n")
print(df)

# Plot Page Views
plt.bar(df["Page"], df["PageViews"])

plt.title("Website Traffic Analysis")
plt.xlabel("Pages")
plt.ylabel("Page Views")

plt.show()
```
